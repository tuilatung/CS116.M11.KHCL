# Đọc ảnh gốc

```python
img = cv.imread('guitar.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

# Nén ảnh và khôi phục sử dụng PCA (K = 64)

```python
pca = PCA(n_components=64)
Z = pca.fit_transform(img)
new_img = pca.inverse_transform(Z)
plt.imshow(new_img, cmap='gray')
```
